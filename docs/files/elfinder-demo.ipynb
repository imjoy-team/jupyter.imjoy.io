# ImJoy elFinder Demo

Try this notebook in https://jupyter.imjoy.io!

```python
from imjoy import api
import imageio
from imjoy_rpc.utils import open_elfinder
import matplotlib.pyplot as plt

async def setup():
    fm = await api.createWindow(
        src="https://jupyter.imjoy.io/elFinder/index.html"
    )

api.export({"setup": setup})
```

```python
from js import fetch

# Download a test image
response = await fetch("https://images.proteinatlas.org/61448/1319_C10_2_blue_red_green.jpg")
data = await response.arrayBuffer()
data = data.to_py()
```

```python
# Write the file to elfinder storage
with open_elfinder("/home/test-image.png", "wb") as f:
    f.write(data)
```

```python
# Read the file from elfinder storage
image_file = open_elfinder("/home/test-image.png", "rb")
img = imageio.imread(image_file) 
img.shape

plt.imshow(img) 
plt.show()
```
